```python
# Create a mapping dictionary for Nation to NOC code
nation_to_noc = {
    'Japan': 'JPN',
    'Brazil': 'BRA',
    'Ukraine': 'UKR',
    'Australia': 'AUS',
    'Russia': 'RUS',
    'Great Britain': 'GBR',
    'Georgia': 'GEO',
    'Italy': 'ITA',
    'Vietnam': 'VIE',
    'Slovenia': 'SLO',
    'Poland': 'POL',
    'South Africa': 'RSA',
    'Canada': 'CAN',
    'France': 'FRA',
    'Belarus': 'BLR',
    'Azerbaijan': 'AZE',
    'Croatia': 'CRO',
    'Finland': 'FIN',
    'Turkey': 'TUR',
    'New Zealand': 'NZL',
    'Mexico': 'MEX'
}

# Replace values in the 'Nation' column with NOC codes
df['Nation'] = df['Nation'].map(nation_to_noc)

```

```python
df['year'] = 2020
```

```python
df["Event"]= "World Cup Baku"
```

```python
# Fill NaN values in the 'ND' column with 0
df['ND'].fillna(0, inplace=True)
```
