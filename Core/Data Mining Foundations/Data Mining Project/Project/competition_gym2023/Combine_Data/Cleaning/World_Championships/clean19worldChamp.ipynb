```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_csv('../../scraping/women/miscellaneous/Women2019_World_Artistic_Gymnastics_Championships.csv')
```

```python
df
```

```python
# Create a new DataFrame from every 5th row (index 4, 9, 14, ...)
removed_rows_df = df.iloc[range(0, len(df), 6)]
```

```python
removed_rows_df
```

```python


# Remove the numbers inside parentheses from the original columns
removed_rows_df['Unnamed: 2'] = removed_rows_df['Unnamed: 2'].str.replace(r'\(\d+\)', '', regex=True)
removed_rows_df['Unnamed: 3'] = removed_rows_df['Unnamed: 3'].str.replace(r'\(\d+\)', '', regex=True)
removed_rows_df['Unnamed: 4'] = removed_rows_df['Unnamed: 4'].str.replace(r'\(\d+\)', '', regex=True)
removed_rows_df['Unnamed: 5'] = removed_rows_df['Unnamed: 5'].str.replace(r'\(\d+\)', '', regex=True)

removed_rows_df
```

```python
removed_rows_df['Year'] = 2019
```

```python
removed_rows_df
```

```python
removed_rows_df.loc[0, 'Rank'] = 1  # Add 1 to the first row
removed_rows_df.loc[6, 'Rank'] = 2  # Add 2 to the second row
removed_rows_df.loc[12, 'Rank'] = 3  # Add 2 to the second row
```

```python
world19champs = removed_rows_df.iloc[:8]
```

```python
world19champs
```

```python

world19champs.rename(columns={'Unnamed: 2': 'vault', 
                                'Unnamed: 3': 'uneven_bars', 
                                'Unnamed: 4': 'balance_beam',
                               'Unnamed: 5': 'floor'}, inplace=True)
```

```python
world19champs.info()
```

```python
# Convert the 'vault' and 'ue' columns from object to float
world19champs['vault'] = pd.to_numeric(world19champs['vault'], errors='coerce')
world19champs['uneven_bars'] = pd.to_numeric(world19champs['uneven_bars'], errors='coerce')
world19champs['balance_beam'] = pd.to_numeric(world19champs['balance_beam'], errors='coerce')
world19champs['floor'] = pd.to_numeric(world19champs['floor'], errors='coerce')
```
