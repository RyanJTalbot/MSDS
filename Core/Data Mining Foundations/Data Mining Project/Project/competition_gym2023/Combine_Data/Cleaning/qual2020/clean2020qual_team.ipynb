```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../../scraping/women/miscellaneous/scrape2020qual_team.csv')
```

```python
df.tail(50)
```

```python

# Select rows 0 and 5 and create a new DataFrame
new_df = pd.concat([df.iloc[0], df.iloc[5], df.iloc[10], df.iloc[15], df.iloc[20],
                   df.iloc[25], df.iloc[30], df.iloc[35], df.iloc[40],
                   df.iloc[45], df.iloc[50], df.iloc[55]], axis=1).transpose()
new_df

# Drop rows 0 and 5 from the original DataFrame
df = df.drop([0, 5,10,15,20,25,30,35,40,45,50,55])

df = df.reset_index(drop=True)

```

```python
new_df
```

```python
df
```

```python
# Extract data within parentheses and store it in a new column
df["Country"] = df['Team'].str.extract(r'\((.*?)\)')

# Remove data within parentheses from the original column
df['Team'] = df['Team'].str.replace(r'\(.*?\)', '')
```

```python
df
```

```python
df['Team'] = df['Team'].apply(lambda x: x.split('(')[0].strip())
df
```

```python
df
```

```python
# Split the 'Vault_Scores' column into 'vault1_score' and 'vault2_score'
df[['vault1_score', 'vault2_score']] = df['vault_score'].str.split('/', expand=True)
df

```

```python
df[['vault2_score', 'vault_avg_score']] = df['vault2_score'].str.split('Avg:', expand=True)
df
```

```python
# Remove leading/trailing whitespace
df['vault2_score'] = df['vault2_score'].str.strip()
df['vault_avg_score'] = df['vault_avg_score'].str.strip()
df
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
df
```

```python
df.info()
```

```python
df['vault_avg_score'] = df['vault_avg_score'].astype(float)
df.info()
```

```python
df['vault2_score'] = df['vault2_score'].astype(float)
df['vault1_score'] = df['vault1_score'].astype(float)
```

```python
df = df.drop(columns=["vault_score"])
df
```

```python
# Remove letters from the column while keeping numbers
df['uneven_rank'] = df['uneven_rank'].str.replace('[A-Za-z]', '', regex=True)
df['bb_rank'] = df['bb_rank'].str.replace('[A-Za-z]', '', regex=True)
df['floor_rank'] = df['floor_rank'].str.replace('[A-Za-z]', '', regex=True)
```

```python
df.info()
```

```python
df['uneven_rank'] = df['uneven_rank'].astype(float)
df['bb_rank'] = df['bb_rank'].astype(float)
df['floor_rank'] = df['floor_rank'].astype(float)
df.info()
```

```python
df["Name"] = df["Team"]
```

```python
df
```

```python
df=df.drop(columns=["Team"])
df
```
