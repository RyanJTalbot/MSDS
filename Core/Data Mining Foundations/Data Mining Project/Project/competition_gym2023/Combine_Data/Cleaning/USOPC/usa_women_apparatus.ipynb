```python
import pandas as pd
import numpy as np
```

```python
bb = pd.read_csv('usopc_women_bb_october.csv')
fx = pd.read_csv('usopc_women_fx_october.csv')
ub = pd.read_csv('usopc_women_ub_october.csv')
vt = pd.read_csv('usopc_women_vt_october.csv')
```

```python
bb=bb[bb['Country']=='USA']
fx=fx[fx['Country']=='USA']
ub=ub[ub['Country']=='USA']
vt=vt[vt['Country']=='USA']
```

```python
bbus=bb.drop(columns=['Country', 'Gender', 'Competition', 
                      'Round', 'Date'])
fxus=fx.drop(columns=['Country', 'Gender', 'Competition', 
                      'Round', 'Date'])
ubus=ub.drop(columns=['Country', 'Gender', 'Competition', 
                      'Round', 'Date'])
vtus=vt.drop(columns=['Country', 'Gender', 'Competition', 
                      'Round', 'Date'])
```

```python
bbus['Name'] = bbus['Name'].str.lower()
fxus['Name'] = fxus['Name'].str.lower()
ubus['Name'] = ubus['Name'].str.lower()
vtus['Name'] = vtus['Name'].str.lower()
```

```python
vtus['Apparatus'] = vtus['Apparatus'].replace(['VT1', 'VT2'], 'VT')
```

```python
bbavg = bbus.groupby(['Name', 'Apparatus']).mean().reset_index()
ubavg = ubus.groupby(['Name', 'Apparatus']).mean().reset_index()
fxavg = fxus.groupby(['Name', 'Apparatus']).mean().reset_index()
vtavg = vtus.groupby(['Name', 'Apparatus']).mean().reset_index()
```

```python
df = pd.concat([bbavg, ubavg, fxavg, vtavg], ignore_index=True)
```

```python
df
```

```python
df=df.drop(columns=['Rank', 'D_Score','E_Score', 'Penalty'])
```

```python
df
```

```python
apparatus_dummies = pd.get_dummies(df['Apparatus'])

# Concatenate the original DataFrame with the new binary columns
df = pd.concat([df, apparatus_dummies], axis=1)

# Multiply each binary column by the corresponding score
for col in apparatus_dummies.columns:
    df[col] = df[col] * df['Score']
```

```python
df=df.drop(columns=['Score', 'Apparatus'])
```

```python
df
```

```python
grouped_df = df.groupby('Name').sum()
grouped_df.reset_index(inplace=True)
grouped_df['Name'].nunique()
df1=grouped_df.copy()
```
