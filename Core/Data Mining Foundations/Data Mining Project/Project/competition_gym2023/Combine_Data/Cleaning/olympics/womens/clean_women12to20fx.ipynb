```python
import pandas as pd
import numpy as np
```

```python
df12 = pd.read_csv('../../../scraping/women/olympics/women2012olympics/Gymnastics_ 2012_Olympics_women\'s_artistic_floor.csv')
df16 = pd.read_csv('../../../scraping/women/olympics/women2016olympics/Gymnastics_ 2016_Olympics_women\'s_artistic_floor.csv')
df20 = pd.read_csv('../../../scraping/women/olympics/womens2020olympics/Gymnastics_ 2020_Olympics_women\'s_artistic_floor.csv')
```

```python
df12['year'] = 2012
```

```python
df16['year'] = 2016
```

```python
df20['year'] = 2020
```

```python
column_mapping = {'Name': 'Gymnast', 
                  'Difficulty': 'D Score','Execution': 'E Score',
                 'Penalty': 'Pen.'}

# Rename the columns using the rename() method
df16.rename(columns=column_mapping, inplace=True)

```

```python
df16['Nation'] = df16['Gymnast'].str.extract(r'\((.*?)\)')
```

```python
df16['Gymnast'] = df16['Gymnast'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
df16.head(1)
```

```python
df20['Nation'] = df20['Gymnast'].str.extract(r'\((.*?)\)')
```

```python
df20['Gymnast'] = df20['Gymnast'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
df20.head()
```

```python
combined_df = pd.concat([df12, df16, df20], axis=0)

combined_df.reset_index(drop=True, inplace=True)
```

```python
df= combined_df.copy()
```

```python
df = df.fillna(0)
```

```python
df
```

```python
df['Nation'].value_counts()
```

```python
# Use the replace method to change values
df['Nation'] = df['Nation'].replace({'China': 'CHN', 'Japan': 'JPN',
                                    'Russia': 'RUS', 'Romania': 'RUM',
                                    'Germany': 'GER', 'Cuba': 'CUB',
                                    'United States': 'USA', 'Ukraine': 'UKR',
                                    'Greece': 'GRE', 'Mexico': 'MEX',
                                    'France': 'FRA', 'Netherlands': 'NED',
                                    'Brazil': 'BRA', "Great Britain": 'GBR',
                                    'South Korea': 'KOR', 'Israel': 'ISR',
                                    "Australia": 'AUS', "Italy": 'ITA'})
```
