```python
# Function to remove parentheses and numbers inside parentheses
def remove_parentheses(entry):
    import re
    return re.sub(r'\([^)]*\)', '', entry)

# List of columns to modify
columns_to_modify = ['vt', 'ub', 'bb', 'fx']

# Loop through the columns and apply the function
for col in columns_to_modify:
    new_df[col] = dff[col].apply(remove_parentheses)
```

```python
# Function to extract content inside parentheses
def extract_country(name):
    import re
    match = re.search(r'\((.*?)\)', name)
    if match:
        return match.group(1)
    else:
        return ''

# Create a new "Country" column by applying the function
dfg['Country'] = dfg['Name'].apply(extract_country)

# Remove words inside parentheses in the "Name" column
dfg['Name'] = dfg['Name'].str.replace(r'\(.*\)', '')
```
