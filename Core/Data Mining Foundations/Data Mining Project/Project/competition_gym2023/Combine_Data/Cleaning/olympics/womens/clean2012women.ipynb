```python
# Function to extract the number inside parentheses
def extract_number_from_parentheses(entry):
    import re
    match = re.search(r'\((\d+)\)', entry)
    if match:
        return match.group(1)
    else:
        return ''

# Apply the function to the "vt" column and create a new "vt_rank" column
dff['fx_rank'] = dff['fx'].apply(extract_number_from_parentheses)
```

```python
new_df = dff.copy()
```

```python
# Function to remove parentheses and numbers inside parentheses
def remove_parentheses(entry):
    import re
    return re.sub(r'\([^)]*\)', '', entry)

# List of columns to modify
columns_to_modify = ['vt', 'ub', 'bb', 'fx']

# Loop through the columns and apply the function
for col in columns_to_modify:
    new_df[col] = dff[col].apply(remove_parentheses)
```
