```python
import pandas as pd
import numpy as np
```

```python
df12 = pd.read_csv('../../../scraping/men/olympics/olympic2012men_final/Gymnastics_ 2012_Olympics_Men\'s_artistic_vault.csv')
df16 = pd.read_csv('../../../scraping/men/olympics/olympic2016men/Gymnastics_ 2016_Olympics_Men\'s_artistic_vault.csv')
df20 = pd.read_csv('../../../scraping/men/olympics/mens202olympics_fianl/Gymnastics_ 2020_Olympics_Men\'s_artistic_vault.csv')
```

```python
df20['year'] = 2020
df16['year'] = 2016
df12['year'] = 2012
```

```python
df20['Nation'] = df20['Gymnast'].str.extract(r'\((.*?)\)')
df20['Gymnast'] = df20['Gymnast'].str.replace(r'\([^)]*\)', '', regex=True)
```

```python
df20.rename(columns={'v1_D_Score': 'vt1_D Score',
                    'v1_E_Score': 'vt1_E Score',
                    'v1_Penalty': 'vt1_Pen.',
                    'Score 1': 'vt1_Vault Score',
                    'v2_D_Score': 'vt2_D Score',
                    'v2_E_Score': 'vt2_E Score',
                    'Penalty':  'vt2_Pen.', 'v2_Score 2': 'vt2_Vault Score',
                    'Position': 'Rank'}, inplace=True)
```

```python
df16.rename(columns={'total': 'Total'}, inplace=True)
```

```python
combined_df = pd.concat([df12, df16, df20], axis=0)

combined_df.reset_index(drop=True, inplace=True)
```

```python
df= combined_df.copy()
```

```python
df = df.fillna(0)
```
