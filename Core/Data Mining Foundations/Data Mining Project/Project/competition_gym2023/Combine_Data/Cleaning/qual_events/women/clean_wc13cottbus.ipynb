```python
import pandas as pd 
import numpy as np
```

```python
bb = pd.read_csv('../../../scraping/women/worldcup13cott/worldcup2013cott_bb.csv')
```

```python
bbq = pd.read_csv('../../../scraping/women/worldcup13cott/worldcup2013cott_bb_qual.csv')
```

```python
bb['round'] = 'final'
```
