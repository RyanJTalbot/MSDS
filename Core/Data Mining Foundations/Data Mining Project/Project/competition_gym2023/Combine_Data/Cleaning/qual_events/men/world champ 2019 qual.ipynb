```python
# Create a MultiIndex for the pivot
df['count'] = df.groupby('Athlete').cumcount() + 1

# Pivot the DataFrame
df_pivoted = df.pivot(index=['Rank', 'Athlete', 'Nation'], columns='count', values=['D', 'E', 'ND', 'Total', 'Average'])

# Flatten the MultiIndex columns
df_pivoted.columns = [f'{col[0]}{col[1]}' if col[1] != '' else f'{col[0]}' for col in df_pivoted.columns]

# Reset index to make 'Rank', 'Athlete', and 'Nation' as regular columns
df_pivoted.reset_index(inplace=True)

df_pivoted
```

```python
df=df_pivoted.copy()
```

```python
df.fillna(0, inplace=True)
```

```python
df=df.drop(columns=['Average2'])
```
