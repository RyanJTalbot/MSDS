# Problem 1

A continuous random variable with cumulative distribution function $F$ has the median value $m$ such that $F(m) = 0.5$. That is, a random variable is just as likely to be larger than its median as it is to be smaller. A continuous random variable with density $f$ has the mode value $x$ for which $f(x)$ attains its maximum. For each of the following four random variables, 

$(i)$ state and graph the density function and then 

$(ii)$ compute the median, mode and mean and show these values on the graph.

$(iii)$ Provide at least one graph for the density function using values of the parameter(s) that you select.  Indicate the median, mode, and mean values on your graph.  (The purpose of this problem is to see the relative locations of the median, mode, and mean for the different random variables).

**a)** $W$ which is uniformly distributed over the interval $[a, b]$.

```r
# Plot a uniform density from a to b
# In this case, we will use 0 to 1
x = seq(0, 1, 0.05)
y = dunif(x)
plot(x=x, y=y, type="l")
# Median is at 1/(a+b), which is 1/(1+0)=1/2 for our plot
med = 1/2
abline(v=med, col="red")
# Mean is at 1/2 as well
mean = 1/2
abline(v=mean, col="blue")
# There is no mode, as all values of f(x) are equal
```

$W$ is uniform on [a,b] so it has density function $f_W(w) = 1/(b-a)$ for all $a \le w \le b$ and $f_W(w) = 0$ for all other values of $w$. 
* We already calculated mean $E[W] = (a+b)/2$. 
* To find the median, we calculate $F(m) = P(W \le m) = \int_a^m \frac{1}{b-a}dw = \frac{m-a}{b-a}$ which we set equal to $1/2$. Solving for $m$ yields the median $m=(a+b)/2$. 
* Since the density function is constant on its nonzero domain, it attains its mode at all values of $w$ for $a \le w \le b$.

**b)** $X$ which is normal with parameters $\mu$ and $\sigma^2$.

```r
# Simulate a standard normal distribution
x = seq(-5, 5, 0.1)
y = dnorm(x)
# Plot the standard normal
plot(x=x, y=y, type="l")
# From the graph, we can see that the mean, median and mode are all equal to mu.
med = 0
mean = 0
mode = 0
abline(v=med, col="red")
abline(v=mean, col="blue")
abline(v=mode, col="green")
```

For $X \sim N(\mu, \sigma^2)$, the median=mean=mode=$\mu$. Doing the actual integration is beyond the scope of this class.

**c)** $Y$ which is exponential with rate $\lambda$.

```r
# Generate a exponential distribution with rate lambda
# For our case, let lambda = 1
lambda = 1
x = seq(0, 20, 0.1)
y = dexp(x, rate=lambda)
# Plot the exponential distribution
plot(x=x, y=y, type="l")

# Mean is 1/lambda.
mean = 1/lambda
abline(v=mean, col="blue")

# Median is ln(2)/lambda.
med = log(2)/lambda
abline(v=med, col="red")

# Mode is at y=0.
mode = 0
abline(v=mode, col="green")
```

$Y$ is exponential with rate $\lambda$ so it has density function $f_Y(y)=\lambda e^{-\lambda y}$ for $y \ge 0$ and $f_Y(y)=0$ for $y<0$.
* The mean can be solved as $E[Y]= \int_0^\infty \lambda y e^{-\lambda y} dy = 1/\lambda$.
* For the median, we have $F(m) = \int_0^m \lambda e^{-\lambda y}dy = 1 - e^{-\lambda m}$, which we set equal to  $1/2$. Solving for $m$, we get $m = (\ln 2)/\lambda$.
* The mode is $\lambda$ and it occurs when $y=0$. We can confirm this on the graph.

# Problem 2

For this problem, we're going to visualize what's happening when we go between different normal distributions.

**Part A)**

Draw at least $10000$ samples from the standard normal distribution $N(0,1)$ and store the results. Make a density histogram of these samples. Set the $x$-limits for your plot to $[-10,10]$ and your $y$-limits to $[0,0.5]$ so we can compare with the plots we'll generate in **Parts B-D**.

```r
samples.1 = rnorm(10000, 0, 1)
hist(samples.1, freq=FALSE, xlim=c(-10, 10), ylim=c(0, 0.5))
```

**Part b)** Now generate $10000$ samples from a $N(2,3)$ distribution and plot a histogram of the results, with the same $x$-limits and $y$-limits. Does the histogram make sense based on the changes to parameters?

Note: Be careful with the parameters for `rnorm()`. It may help to check the documentation.

```r
samples.2 = rnorm(10000, mean=2, sd=sqrt(3))
hist(samples.2, freq=FALSE, xlim=c(-10, 10), ylim=c(0, 0.5))
```

We can see that the histogram is now centered around $2$, which makes sense because that is the new mean value. The histogram is also much wider, which also makes sense as we increased the variance (thus spread) of the data.

**Part c)**

Suppose we are only able to sample from the standard normal distribution $N(0,1)$. Could we take those samples and perform a simple transformation so that they're samples from $N(2,3)$? Try this, and plot another histogram of the transformed data, again with the same axes. Does your histogram based of the transformed data look like the histogram from **Part B**?

```r
samples.1.transformed = samples.1 * sqrt(3) + 2
hist(samples.1.transformed, freq=FALSE, xlim=c(-10, 10), ylim=c(0, 0.5))
```

We can see that the two histograms do match! The samples agree with the theory.

**Part d)** 

But can you go back the other way? Take the $N(3,4)$ samples from **Part B** and transform them into samples from $N(0,1)$? Try a few transformations and make a density histogram of your transformed data. Does it look like the plot of $N(0,1)$ data from **Part A**?

```r
samples.2.transformed = (samples.2-2)/sqrt(3)
hist(samples.2.transformed, freq=FALSE, xlim=c(-10, 10), ylim=c(0, 0.5))
```

Again, we see that the histograms are basically the same. This should cement the idea that any normal distribution can be transformed into the standard normal, and the standard normal can be transformed to any other normal distribution.
