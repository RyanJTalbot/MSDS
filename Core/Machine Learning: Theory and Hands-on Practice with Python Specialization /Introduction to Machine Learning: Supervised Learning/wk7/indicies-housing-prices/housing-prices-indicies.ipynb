```python

# Specify the path to your Excel file
file_path = "snb-data-plimoinreg.xlsx"  # Replace with your file path

# Read the Excel file and get the sheet names to verify the sheets available
sheet_names = pd.ExcelFile(file_path).sheet_names
print(f"Available Sheets: {sheet_names}")

# Read the second sheet by index (index starts at 0)
second_sheet_df = pd.read_excel(file_path, sheet_name=sheet_names[1])


```
