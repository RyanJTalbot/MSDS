```python
merged_df = pd.merge(df, int, on='Year', how='inner')
```

```python
df = merged_df.copy()
```

```python
# Convert Inflation Rate from percentage to decimal
df['Inflation Rate Decimal'] = df['Inflation Rate YOY, From Previous Dec. (%)'] / 100

# Initialize the Cumulative Inflation Factor
# Start with 1 for the base year (1990)
df['Cumulative Inflation Factor'] = (1 + df['Inflation Rate Decimal']).cumprod()

# Adjust the Cumulative Inflation Factor so that the base year is 1
# Since the base year should not be adjusted by its own inflation rate
df['Cumulative Inflation Factor'] = df['Cumulative Inflation Factor'] / (1 + df['Inflation Rate Decimal'])

# Replace NaN with 1 for the base year
df.loc[df['Year'] == df['Year'].min(), 'Cumulative Inflation Factor'] = 1
```

```python
fig, ax1 = plt.subplots(figsize=(12, 7))

color = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Cost per Mile (USD)', color=color)
sns.lineplot(data=df, x='Year', y='Cost per Mile', marker='o', color=color, label='Cost per Mile', ax=ax1)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_xticks(df['Year'])

ax2 = ax1.twinx()  # Instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Inflation Rate YOY (%)', color=color)  # We already handled the x-label with ax1
sns.lineplot(data=df, x='Year', y='Inflation Rate YOY, From Previous Dec. (%)', marker='s', color=color, label='Inflation Rate YOY', ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # Otherwise the right y-label is slightly clipped
plt.title('Cost per Mile and Inflation Rate Over Years (1990-1995)')
plt.show()
```
