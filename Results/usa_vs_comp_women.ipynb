```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Score', y='Country', data=df, color='lightblue', orient='h')
plt.title('Algorithm Qualification Scores by Country')
plt.xlabel('Score')
plt.ylabel('Country')
plt.xlim(158, 170)  # Set x-axis limit starting from 150

# Show plot
plt.tight_layout()
plt.show()
```
