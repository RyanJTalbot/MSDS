```r
#replicate
set.seed(110)
sims = 1000

#generate the r.v.'s
X = rnorm(sims)
Y = rnorm(sims)

#set a path for Z
Z = rep(NA, sims)

#run a loop to calculate Z
for(i in 1:sims){
  
  #flip to see if Z takes on X or Y
  flip = runif(1)
  
  #Z takes on X
  if(flip <= .99){
    Z[i] = X[i]
  }
  
  #Z takes on Y
  if(flip > .99){
    Z[i] = Y[i]
  }
}

#set 2x2 graphing grid
par(mfrow = c(2,2))

#plot 4 histograms
hist(X, main = "X", col = rgb(1, 0, 0, 1/4), xlab = "")
hist(Y, main = "Y", col = rgb(0, 1, 0, 1/4), xlab = "")
hist(Z, main = "Z", col = rgb(0, 0, 1, 1/4), xlab = "")
hist(rnorm(sims), main = "N(0, 1)", col = "gray", xlab = "")
```

```r
#set 1x2 graphing grid
par(mfrow = c(1,2))

#histograms do not match
hist(X - Z, main = "X - Z", col = rgb(1, 0, 1, 1/4),
     xlab = "")
hist(rnorm(sims), main = "N(0, 1)", col = "gray",
     xlab = "")
```
